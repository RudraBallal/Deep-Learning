```python
import cv2
```

```python
mario = cv2.imread('mario.jpg')
mario
```

```python
mario.shape
```

```python
import matplotlib.pyplot as plt
plt.imshow(mario)
plt.show()
```

### Showing image as RGB

```python
mario_rgb = cv2.cvtColor(mario, cv2.COLOR_BGR2RGB)
mario_rgb
```

```python
plt.imshow(mario_rgb)
plt.show()
```

### Getting RGB channels for above image

```python
r, g, b = cv2.split(mario_rgb)
```

```python
r
```

```python
r.shape
```

```python
# Red Channel
plt.imshow(r, cmap='Reds')
plt.show()
```

```python
# Green channel
plt.imshow(g, cmap='Greens')
plt.show()
```

```python
# Blue channel
plt.imshow(b, cmap="Blues")
plt.show()
```

### Reading Grayscale Image

```python
mario_gray = cv2.imread('mario.jpg', cv2.IMREAD_GRAYSCALE)
mario_gray
```

```python
mario_gray.shape
```

```python
plt.imshow(mario_gray, cmap='gray')
plt.show()
```
